# CSV Cleaning

```python
df_2 = pd.read_csv('/content/data_2.csv')
df_4 = pd.read_csv('/content/data_4.csv')
df_6 = pd.read_csv('/content/data_6.csv')
df_7 = pd.read_csv('/content/data_7.csv')
df_9 = pd.read_csv('/content/data_9.csv')
df_16 = pd.read_csv('/content/data_16.csv')
```

```python
counties = list(df_2['savivaldybe'].unique())
del counties[counties.index('Iš viso (asmenų)')]
```

```python
df_2 = df_2[df_2['savivaldybe'].isin(counties)]
df_4 = df_4[df_4['savivaldybe'].isin(counties)]
df_6 = df_6[df_6['savivaldybe'].isin(counties)]
df_7 = df_7[df_7['savivaldybe'].isin(counties)]
df_9 = df_9[df_9['savivaldybe'].isin(counties)]
df_16 = df_16[df_16['savivaldybe'].isin(counties)]
```

```python
df = df_2.copy(deep=True)
```

```python
df =(
    df.merge(df_4, on=['savivaldybe', 'mokslo_metai'])
      .merge(df_6, on=['savivaldybe', 'mokslo_metai'])
      .merge(df_7, on=['savivaldybe', 'mokslo_metai'])
      .merge(df_9, on=['savivaldybe', 'mokslo_metai'])
      .merge(df_16, on=['savivaldybe', 'mokslo_metai'])
)
```

```python
columns = list(df.columns)
columns.remove('savivaldybe')
columns.remove('mokslo_metai')
columns = ['savivaldybe', 'mokslo_metai'] + columns
```

```python
df[columns].to_csv('padagogu_datasetas.csv', index=False)
```

# Similarity Calculation

```python
import pandas as pd

df = pd.read_csv('/content/padagogu_datasetas.csv')

df.head(5)
```

```python
import itertools
import numpy as np
```

```python
cols = list(df.columns)
cols.remove('savivaldybe')
cols.remove('mokslo_metai')
```

```python
cols = ['mokytoju_skaicius', 'ekspertu_skaicius', 'mokiniu_skaicius', 'mokiniu_mokytojui_sk', 'mokytoju_skaicius_ne_vienoje_mokykloje', 'moko_daugiau_nei_1_dalyko', 'pensijinio_amziaus', 'stazas_iki_2_metu_proc']
```

```python
new_data = []
```

```python
def calculate_similarity(row1, row2):
  mokyt_sk = round(abs(row1['mokytoju_skaicius'] - row2['mokytoju_skaicius']) / max(row1['mokytoju_skaicius'], row2['mokytoju_skaicius']), 4)
  mok_sk = round(abs(row1['mokiniu_skaicius'] - row2['mokiniu_skaicius']) / max(row1['mokiniu_skaicius'], row2['mokiniu_skaicius']), 4)
  ne_1_mok = round(abs(row1['mokytoju_skaicius_ne_vienoje_mokykloje'] - row2['mokytoju_skaicius_ne_vienoje_mokykloje']) / max(row1['mokytoju_skaicius_ne_vienoje_mokykloje'], row2['mokytoju_skaicius_ne_vienoje_mokykloje']), 4)
  avg_all = round(np.mean([mokyt_sk, mok_sk, ne_1_mok]), 4)
  return mokyt_sk, mok_sk, ne_1_mok, avg_all

for year, group in df.groupby('mokslo_metai'):
  for (idx1, row1), (idx2, row2) in itertools.product(group.iterrows(), repeat=2):
    mokyt_sk, mok_sk, ne_1_mok, avg_all = calculate_similarity(row1, row2)
    new_data.append([row1['savivaldybe']]+list(row1[cols])+[row2['savivaldybe']]+list(row2[cols])+[year, mokyt_sk, mok_sk, ne_1_mok, avg_all])
    #new_data.append([row1['savivaldybe'], row2['savivaldybe'], year, mokyt_sk, mok_sk, ne_1_mok, avg_all]+list(row1[cols]))
new_df = pd.DataFrame(new_data, columns=['savivaldybe_1']+cols+['savivaldybe_2']+[f"{i}_2" for i in cols]+ ['mokslo_metai', 'mokyt_simil', 'mok_simil', 'ne_1_mok_simil', 'avg_simil'])
```
